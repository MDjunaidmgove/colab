```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
df = sns.load_dataset('titanic')
num_cols = df[['age','fare','pclass']].dropna()
cov_matrix = num_cols.cov()
print("coveriance:\n", cov_matrix)
corr_matrix = num_cols.corr()
print("\ncorrelation:\n", corr_matrix)
```

```python
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
       'Month':['jan','feb','mar','apr'],
       'product_A':[30,25,45,44],
       'product_B':[25,33,44,21],
       'product_c':[10,23,41,22]}
df = pd.DataFrame(data)
df.set_index('Month').plot(kind='bar', stacked=True)
plt.ylabel('sales')
plt.title('monthly sales by product')
plt.show()
```

#**eigen values and eigen vectors**

```python
import numpy as np
A = np.array([[2, -1], [-1, 3]])
eigenvalues, eigenvectors = np.linalg.eig(A)
print("Eigenvalues:")
print(eigenvalues)
print("\nEigenvectors:")
print(eigenvectors)
```

#***Using* SciPy:**

```python
import numpy as np
from scipy.linalg import eig
A = np.array([[2, -1], [-1, 3]])
eigenvalues, eigenvectors = eig(A)
print("Eigenvalues:")
print(eigenvalues)
print("\nEigenvectors:")
print(eigenvectors)
```

```python
import numpy as np
from scipy.linalg import eigvals
A = np.array([[2, -1], [-1, 3]])
eigenvalues = eigvals(A)
print("Eigenvalues:")
print(eigenvalues)4
```
